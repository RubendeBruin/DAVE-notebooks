# Making the hook asset

Learning objectives:

- hinge points
- saving a scene


```python
from DAVE import *
from DAVE.jupyter import *
```

```python
s = Scene()
```

Upper block is fixed to the world

```python
s.new_rigidbody('upper_block', mass = 30, cog = (0,0,1), fixed=True)
s.new_visual('uh_visual', parent='upper_block', path='upper_block.obj')
show(s, sea=False)
```

The origin of the upper-block is in the pivot-axis.


The lower-block must be
- attached to the upper block.
- be able to pivot about the pivot-axis of the upper block.

Axis (and rigid-bodies) can be allowed to pivot about their origin. This can be done by setting the "fixed"-property for the dof to False. In this case we want the lower block to be able to pivot about the X-axis of the upper-block. So the fixed-property will become True, True,True (no translation) False, True , True (only rotation about X is not fixed).

To make the lower-block pivot about the correct point we need to use the pivot of the upper block as origin for the lower block. This is (0,0,0).



```python
s.new_rigidbody(name='lower_block',
               parent = 'upper_block',
               fixed = (True, True, True, False, True, True),
               mass = 15,
               cog = (0,0,-0.75))

s.new_visual('lb_visual', parent='lower_block', offset = (0,0,-1.5), path = 'lower_block.obj')
show(s, sea=False)
```

The hook needs to be attached to the pivot of the lower block.
The hook needs to be able to rotate about the y-axis.

the "fixed" property should be (True, True, True, True, False, True)
the origin of the lower hook should be (0,0, -1.5)

```python
s.new_rigidbody(name='hook4p',
           parent = 'lower_block' ,
           position = (0,0,-1.5),
           mass = 80,
           cog = (0,0, -1.3),
           fixed = (True, True,True,True,False,True) )

s.new_visual(name='Visual',
            parent='hook4p',
            path='hook4p.obj')

```

```python
show(s, sea=False, camera_pos = (20,10,0), cog_scale = 0)
```

Add the pois for the prongs

```python
s.new_point(name='prong1',
          parent='hook4p',
          position=(1.2,
                    0.0,
                    -1.3))
s.new_point(name='prong2',
          parent='hook4p',
          position=(0.0,
                    1.2,
                    -1.3))
s.new_point(name='prong3',
          parent='hook4p',
          position=(-1.2,
                    0.0,
                    -1.3))
s.new_point(name='prong4',
          parent='hook4p',
          position=(0.0,
                    -1.2,
                    -1.3))
```

A scene can be saved for re-use on other models using "save_scene".

```python
s.save_scene('hook4p')
```

# Testing the rotation axis

Let's see if the rotation axis have been defined correcly. This is easy to do by pulling the hook away by a force, solving and looking at the result.

```python
s.new_force('testing force', parent = 'prong4', force=(300,300,0))
s.solve_statics()
```

```python
show(s, sea=False, cog_scale=0)
```

Looks as expected!
