# GZ-curves

This example shows how to create a GZ-curve for a free-floating vessel using the imposed-displacement method.

Besides the standard imports, import the marine package:

```python
from DAVE import *
from DAVE.jupyter import *
from DAVE.marine import *
```

Create a vessel with buoyancy based on the cube-with-bevel mesh

```python
s = Scene()

s.new_rigidbody(name='vessel',
                mass=20.0,
                cog=(0.0,
                     0.0,
                     1.0))
# code for buoyancy
s.new_buoyancy(name='buoyancy',
          parent='vessel')
s['buoyancy'].trimesh.load_obj(s.get_resource_path('cube_with_bevel.obj'), scale = (10.0,4.0,1.0))
```

Just to check if the model looks correct:

```python
s.solve_statics(silent=True)

from DAVE.gui import *
Gui(s)
```

Creating a GZ-curve is a single command.

This "displacement driven" method works when a vessel does not have a parent.

It will temporarily fix the vessel under an angle, calculate statics and then evaluate the heeling force on the fix as restoring moment.

```python
GZcurve_DisplacementDriven(s, 'vessel'); # <=== the ; at the end suppresses the numerical output
```

By default the restoring moment is plotted. If the GZ value is needed then a displacement needs to be provided.

Also, since the vessel did not trim, we might as well fix it.

```python
GZcurve_DisplacementDriven(s, 'vessel', displacement_kN = 20 * 9.81, allow_trim=False);
```

A range and number of steps may be provided as well:

```python
data = GZcurve_DisplacementDriven(s, 'vessel', displacement_kN = 20 * 9.81, 
                           minimum_heel=-20, maximum_heel=30, steps=101, 
                           allow_trim=False);
```
