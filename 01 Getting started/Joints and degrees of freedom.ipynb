# Joints using degrees of freedom (1)

The preferred method of constructing joints such a sliders, hinges and ball-joints is by using the degrees of freedom.

In virtual-float every axis or rigid-body node has six degrees of freedom:

1. Linear motion in X
2. Linear motion in Y
3. Linear motion in Z
4. Rotation motion about X
5. Rotation motion about Y
6. Rotation motion about Z

All of these can be individually set to FIXED or FREE. If a dof (degree of freedom) is FIXED then then value of this dof is not allowed to change. If it is not fixed then the solver will change the value of the dof such that the resulting force or moment on this dof is zero.

# Play-time !

The following cell creates the demo-scene. In the two cells after that you can experiment with different initial positions and fixed/free degrees of freedom:

```python
s = Scene()
# code for Axis
s.new_rigidbody(name='Body',
           mass = 1,
           cog = (3,0,0) ,
           fixed = False )
# code for Poi_on_axis
s.new_point(name='Poi_on_body',
          parent='Body',
          position=(5.0,
                    0.0,
                    0.0))

s.new_visual(name='body_visual',
            parent = 'Body',
            path = 'cube.obj',
            scale = (6,0.5,0.5),
            offset = (3,0,0))

# code for global_poi
s.new_point(name='global_poi',
          position=(7.0,
                    0.0,
                    3.0))
# code for Cable
s.new_cable(name='Cable',
            endA='Poi_on_body',
            endB='global_poi',
            length=2,
            EA=1000.0)
# code for Global_origin
s.new_axis(name='Global_origin',
           position=(0.0,
                     0.0,
                     0.0),
           rotation=(0.0,
                     0.0,
                     0.0),
           fixed = True )
```

```python
s['Body'].position = (0,0,0)   # initial position
s['Body'].rotation = (0,0,0)   # intial rotation
s['Body'].fixed = (True, True, True,    # x,y and z translation
                   False, False, False) # rotations about x,y and z axis

print('Initial condition:')

show(s, sea = False, camera_pos = (2.9504534625106804, -20.54940468756877, 4.00304171717938), lookat = (4.092818364882734, 4.424212834655461, 1.3559591275879552))
```

```python
print('Initial position = {} {} {}'.format(*s['Body'].position))
print('Initial rotation = {} {} {}'.format(*s['Body'].rotation))
print('X-translation allowed: {}'.format(s['Body'].fixed[0]))
print('Y-translation allowed: {}'.format(s['Body'].fixed[1]))
print('Z-translation allowed: {}'.format(s['Body'].fixed[2]))
print('X-rotation allowed: {}'.format(s['Body'].fixed[3]))
print('Y-rotation allowed: {}'.format(s['Body'].fixed[4]))
print('Z-rotation allowed: {}'.format(s['Body'].fixed[5]))

s.solve_statics()

print('Results in:')
print('Solved position = {} {} {}'.format(*s['Body'].position))
print('Solved rotation = {} {} {}'.format(*s['Body'].rotation))

print('\n\n Solved position')

show(s, sea = False, camera_pos = (2.9504534625106804, -20.54940468756877, 4.00304171717938), lookat = (4.092818364882734, 4.424212834655461, 1.3559591275879552))
```

# Rotations

Rotations are defined using the rotation vector: [https://en.wikipedia.org/wiki/Axis%E2%80%93angle_representation#Rotation_vector]

Note:
Fixing one of the components of the rotation vector to a non-zero value results in very non-intuitive results. I can't even explain what it would mean physically. Therefore this is not allowed.
Fixed components of the rotation vector are set to zero when at least one of the other components is free.

If you want to make complex rotations then it is often much easier to simply stack multiple axes on top of eachother.


# Parents

Besides that every axis or rigid-body node *can* have a parent node. If a node has a parent then the degrees of freedom of that node are relative to that parent. If a node does not have a parent then the degrees of freedom are relative to the global axis system.
